## Upload file to Assembly Ai's server

```python
def read_file(filename, chunk_size=5242880):
    with open(filename, 'rb') as _file:
        while True:
            data = _file.read(chunk_size)
            if not data:
                break
            yield data
 
headers = {'authorization': api_key}
response = requests.post('https://api.assemblyai.com/v2/upload',
                         headers=headers,
                         data=read_file(filename))

audio_url = response.json()['upload_url']

print('File successfully uploaded to AssemblyAI')
```

## Transcribe audio file

```python

endpoint = "https://api.assemblyai.com/v2/transcript"

json = {
  "audio_url": audio_url
}

headers = {
    "authorization": api_key,
    "content-type": "application/json"
}

transcript_input_response = requests.post(endpoint, json=json, headers=headers)

print('Transcribing file')
```

## Fetch transcript ID

```python
transcript_id = transcript_input_response.json()["id"]
print("Transcript ID: ",transcript_id)
```

## Retrieve transcription results


```python
endpoint = f"https://api.assemblyai.com/v2/transcript/{transcript_id}"
headers = {
    "authorization": api_key,
}
transcript_output_response = requests.get(endpoint, headers=headers)

print(transcript_output_response.json()["status"])
```

## Auto-correct transcribed text

```python
#Convert json file to human readable text
result = (transcript_output_response.json()["text"])

result = result.upper()


#Auto-correct

if "THREE" in result:
            result = result.replace("THREE","3")
        
if "FOUR" or "FOR" in result:
            result = result.replace("FOUR" or "FOR","4")
        
if "FIVE" in result:
            result = result.replace("FIVE","5")

if "PI" in result:
            result = result.replace("PI","5")

if "SIX" in result:
            result = result.replace("SIX","6")

if "SEX" in result:
            result = result.replace("SEX","6")

if "EIGHT" in result:
            result = result.replace("EIGHT","8")

if "NINE" in result:
            result = result.replace("NINE","9")

if "BE" in result:
            result = result.replace("BE","B")

if "AND" in result:
            result = result.replace("AND","N")
        
if "TE" in result:
            result = result.replace("TE","T")

if "EX" in result:
            result = result.replace("EX","X")
        
#Remove whitespace
if " " in result:
    result = result.replace(" ","")

#Remove full stops
if "." in result:
    result = result.replace(".","")
    
#Remove hyphen
if "-" in result:
    result = result.replace("-","")
    
#Remove comma
if "," in result:
    result = result.replace(",","")
```
